```python
import json
import pprint as pp

with open('air.json', encoding='utf-8') as f:
    data = json.load(f)

sort = sorted(data['航空公司'], key=lambda x: int(x['員工數目'].replace(',', '')))
print("航司人數由低排到高：")
pp.pprint(sort, sort_dicts=False)
```

```python
import json

# 讀取 JSON 文件
with open('air.json', encoding='utf-8') as f:
    data = json.load(f)

# 輸入航司關鍵字
x = input("請輸入航司關鍵字：")

# 遍歷航空公司數據，查找包含關鍵字的航司
for company in data['航空公司']:
    if x in company['航空公司']:
        print(f"{company['航空公司']}的收入為 {company['收入 (億)']} (億)元")
```
