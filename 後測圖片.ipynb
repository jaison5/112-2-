```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('adult.csv')

income = data['Income']
income = data['Age']

plt.figure(figsize=(10, 6))

plt.hist(income, bins=10, edgecolor='k')

plt.xlabel('age')
plt.ylabel('income')
plt.title('Income Distribution')

plt.show()
```
